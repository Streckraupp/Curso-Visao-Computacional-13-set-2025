```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load an image from file

imagepath = "images/jardimdejacares.jpg"
image = cv2.imread(imagepath)
if image is None:
	raise FileNotFoundError(f"Image {imagepath} not found or could not be loaded.")
rbg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(rbg, cmap=None)
plt.show()
```

```python
# Now test what happens when we do not use "cv2.COLOR_BGR2RGB"

plt.axis('off')
plt.imshow(image, cmap=None) # type: ignore
plt.show()
```

```python
height, width, channels = image.shape
center = (width // 2, height // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, angle=45, scale=1.0)
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
plt.axis('off')
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB), cmap=None)
plt.show()
```
